```python
import copy
```

```python
class Helper:
    
    def __init__(self):
        self.direcao = {0:"Esquerda", 1:"baixo", 2:"Direita",3:"Cima"}

    def printNode(self, node):
        for n in node:
            print(n)
            
    def printNodeFilhos(self, listnode):
        strconcat = "| "
        for i, dire in self.direcao.items():
            strconcat += str(i)+"="+str(dire)+" | "
        print(strconcat)

        for i in range(3):
            strconcat = "|"
            for esc in listnode:
                if esc is not None:
                    strconcat += str(esc[i]) + "|"
                else:
                    strconcat += "[  ,  ,  ] |"

            print(strconcat)
    
    def comparar(self, a, b):
        return len([x for x, y in zip(a,b) if x != y]) == 0
        
helper = Helper()
```

```python
class Node:
    def __init__(self, node):
        self.node = node
        
    def getFilhos(self, direcao=None):
        tmp = []
        if direcao is None:
            tmp.append(self.goEsquerda())
            tmp.append(self.goBaixo())
            tmp.append(self.goDireita())
            tmp.append(self.goCima())
        elif direcao == 0:
            tmp = self.goEsquerda()
        elif direcao == 1:
            tmp = self.goBaixo()
        elif direcao == 2:
            tmp = self.goDireita()
        elif direcao == 3:
            tmp = self.goCima()
        return tmp
        pass
    
    def goBaixo(self):
        newnode = copy.deepcopy(self.node)
        flag = False
        for i in range(1, len(newnode)):
            for j in range(len(newnode[0])):
                if newnode[i-1][j] == 0:
                    newnode[i-1][j] = newnode[i][j]
                    newnode[i][j] = 0
                    flag = True
                    break
            if flag:
                break
        if not flag:
            return None
        return newnode
    
    def goCima(self):
        newnode = copy.deepcopy(self.node)
        flag = False
        for i in range(0, len(newnode)-1):
            for j in range(len(newnode[0])):
                if newnode[i+1][j] == 0:
                    newnode[i+1][j] = newnode[i][j]
                    newnode[i][j] = 0
                    flag = True
                    break
            if flag:
                break
        if not flag:
            return None
        return newnode
    
    def goDireita(self):
        newnode = copy.deepcopy(self.node)
        flag = False
        for i in range(len(newnode)):
            for j in range(1,len(newnode[0])):
                if newnode[i][j-1] == 0:
                    newnode[i][j-1] = newnode[i][j]
                    newnode[i][j] = 0
                    flag = True
                    break
            if flag:
                break
        if not flag:
            return None
        return newnode
    
    def goEsquerda(self):
        newnode = copy.deepcopy(self.node)
        flag = False
        for i in range(len(newnode)):
            for j in range(0,len(newnode[0])-1):
                if newnode[i][j+1] == 0:
                    newnode[i][j+1] = newnode[i][j]
                    newnode[i][j] = 0
                    flag = True
                    break
            if flag:
                break
        if not flag:
            return None
        return newnode
```

```python
teste = [
    [1,2,3],
    [4,0,5],
    [6,7,8]
]
helper.printNode(teste)

espander = Node(teste)
print("------------------------")
helper.printNodeFilhos(espander.getFilhos())
```

```python
def testaJogadas(inicial, final, lista):
    if lista is None:
        print("Não existe inicial - final")
        return
    tmp = Node(inicial)
    print("##INICIO##")
    helper.printNode(inicial)
    final = [[]]
    for i in lista:
        tmp = tmp.getFilhos(i)
        print("| "+helper.direcao[i]+" |")
        helper.printNode(tmp)
        final = tmp
        tmp = Node(tmp)
        
    if helper.comparar(final, final):
        print("CHEGOU!!")
    else:
        print("NAO CHEGOU!!")
    
```
